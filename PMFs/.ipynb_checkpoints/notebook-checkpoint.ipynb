```python
import mdtraj as md
import itertools, math
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

kB = 1.38064852*10**-23
T = 300
fig, axs = plt.subplots(2, 2)
ion_names = ['K','NA']
for macrocycle in ['0','1']:
    for solvent in ['a','w']:
        for x,ion in enumerate(['k','n']):
            prefix = f'{macrocycle}_{solvent}_{ion}'
            traj = md.load(f'{prefix}.trr',top=f'{prefix}.gro')
            carbonyl_o_indices = [a.index for a in traj.topology.atoms if a.name == 'O'][:6]
            ion_indices = [a.index for a in traj.topology.atoms if a.name == ion_names[x]]
            pairs = itertools.product(carbonyl_o_indices, ion_indices)
            ion_O_distances = np.concatenate(md.compute_distances(traj, pairs))
            counts, distances = np.histogram(ion_O_distances, 1000)
            counts = [x+1 for x in counts]
            pmf = [-kB*T*math.log(counts[i]) + kB*T*2*distances[i] for i in range(len(counts))]
            
            plt.figure(2, figsize=(10,5))
            plt.plot(distances[1:],pmf,label=prefix)
            
            if macrocycle == '0' and solvent == 'a':
                axs[0,0].plot(distances[1:],pmf, color='r')
            if macrocycle == '1' and solvent == 'a':
                axs[0,0].plot(distances[1:],pmf, color='b')
            axs[0,0].set_title('PMF comparing macrocycles in acetonitrile')
            if ion == 'n' and solvent == 'w':
                axs[0,1].plot(distances[1:],pmf, color='r')
            if ion == 'k' and solvent == 'w':
                axs[0,1].plot(distances[1:],pmf, color='b')
            axs[0,1].set_title('PMF comparing ions in water')
            if ion == 'n' and solvent == 'a':
                axs[1,0].plot(distances[1:],pmf, color='r')
            if ion == 'k' and solvent == 'a':
                axs[1,0].plot(distances[1:],pmf, color='b')
            axs[1,0].set_title('PMF comparing ions in acetonitrile')
            if macrocycle == '0' and solvent == 'w':
                axs[1,1].plot(distances[1:],pmf, color='r')
            if macrocycle == '1' and solvent == 'w':
                axs[1,1].plot(distances[1:],pmf, color='b')
            axs[1,1].set_title('PMF comparing macrocycles in water')
plt.xlim(0.1,1)
plt.legend()
fig.tight_layout()
plt.savefig('pmf.pdf')
fig.savefig('pmfs.pdf')
plt.close()
```
